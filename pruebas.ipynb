```python
def rebin(arr, new_shape):
    shape = (new_shape[0], arr.shape[0] // new_shape[0],
             new_shape[1], arr.shape[1] // new_shape[1])
    return arr.reshape(shape).mean(-1).mean(1)
```

```python
import numpy as np
def scale(img, nR, nC):
  nR0 = len(img)     
  nC0 = len(img[0])  
  new = [[ img[int(nR0 * r / nR)][int(nC0 * c / nC)] for c in range(nC)] for r in range(nR)]
  return np.array(new)
```

```python
c = scale(img, 1000, 1000)
```
